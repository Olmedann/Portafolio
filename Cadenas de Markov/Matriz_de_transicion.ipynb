```python
import pandas as pd
import numpy as np

gen14 = pd.read_excel("gen14.xlsx")
gen15 = pd.read_excel("gen15.xlsx")
gen1601 = pd.read_excel("gen1601.xlsx")
gen1602 = pd.read_excel("gen1602.xlsx")
gen16 = gen1601 + gen1602
gen17 = pd.read_excel("gen17.xlsx")
gen18 = pd.read_excel("gen18.xlsx")
gen19 = pd.read_excel("gen19.xlsx")
gen20 = pd.read_excel("gen20.xlsx")
gen21 = pd.read_excel("gen21.xlsx")
gen22 = pd.read_excel("gen22.xlsx")

archivos = [gen14,
            gen15,
            gen16,
            gen17,
            gen18,
            gen19,
            gen20,
            gen21,
            gen22]

def obtener_estados(tabla):
    matriz = np.zeros((12,12))
    #ponemos 1 a los estados absorbentes
    matriz[11,11] = 1
    matriz[10,10] = 1
    # semestres
    diagonal = [tabla[columna][columna - 1] for columna in range(2,11)]
    #llenamos los datos en la matriz
    for i in range(1,10):
        matriz[i-1][i] = diagonal[i-1]
    #alumnos rezagados
    rezagados = [p-a for p,a in zip(tabla["Poblacion"][1:11],diagonal)]
    #llenamos los datos en la matriz
    for i,x in enumerate(rezagados):
        matriz[i][i] = x
    #alumnos desertores
    desertores = [tabla["Poblacion"][i-1] - tabla["Poblacion"][i] for i in range(1,10)]
    #escribimos
    for i,x in enumerate(desertores):
        matriz[i][11] = x
    #Egresados
    numero_egresados = tabla["Egresados"].iat[-1]
    matriz[9][10] = numero_egresados
    matriz[9][9] = numero_egresados - tabla["Egresados"].iat[9]
    matriz[9][11] = tabla["Poblacion"][8] - tabla["Poblacion"][9]
    return np.nan_to_num(matriz)

def obtener_matriz(tabla):
    for i in range(12):
        if sum(tabla[i]) > 0:
            tabla[i] = 1/(sum(tabla[i])) * tabla[i]
    return np.nan_to_num(tabla)

def matriz_transicion(tabla):
    return obtener_matriz(obtener_estados(tabla))

# lista que contiene la matriz de conteo de cada generacion
E = list(map(obtener_estados,archivos))

```

Calculo de matrices sin considerar la pandemia

```python
def no_pandemia(gen,semestre,final = None):
  m = obtener_estados(gen)
  m[semestre - 1:final,:] = 0
  return m
npd15 = no_pandemia(gen15,10)
npd16 = no_pandemia(gen16,8)
npd17 = no_pandemia(gen17,6,10)
npd18 = no_pandemia(gen18,4,8)
npd19 = no_pandemia(gen19,2,6)

nopd = sum([obtener_estados(gen14),
            npd15,
            npd16,
            npd17,
            npd18,
            npd19])
pd.DataFrame(obtener_matriz(nopd)).to_excel("matriz no pandemia.xlsx")
P = mc.MarkovChain(obtener_matriz(nopd),nombres)
pd.DataFrame(P.absorption_probabilities(),columns=list(range(1,11))).to_excel("abs prob no pandemia.xlsx")
pd.DataFrame(P.mean_absorption_times()).to_excel("mean time absorption.xlsx")
```
